```python
from datasets import load_dataset, Dataset, DatasetDict
import string 
import unicodedata
import re
import librosa
import os

def preprocess_text(text: str):
    translator = str.maketrans(' ', ' ', r"""!"#$&'()*+,-./:;<=>?@[\]^_`{|}~""")
    text = text.translate(translator)
    text = text.lower()
    text = unicodedata.normalize("NFKC", text)
    text = text.replace('\xa0', '')
    text = re.sub(r"\s{2,}", ' ',text)
    return text

def get_array_from_path(audio_path):
    try:
        root_path = '/mnt/wsl/PHYSICALDRIVE0p1/toan/ASR_Label/audios'
        y, _ = librosa.load(os.path.join(root_path, audio_path), sr=16000, mono=True)
        return y
    except:
        print(audio_path)
        return None
```

```python
df['arr'] = df['path'].apply(get_array_from_path)
```

```python
df_ = df[df['arr'].notnull()]
df_
```
